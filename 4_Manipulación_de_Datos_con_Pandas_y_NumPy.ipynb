**Creacion y manipulacion de Dataframes**

**Combinacion de dataframes**

```python
# Aplicar una función a una columna
df2['Doble_Edad'] = df2['Edad'].apply(lambda x: x * 2)
df2
```

1. **Concatenación**: para combinar DataFrames a lo largo de un eje. Es útil cuando quieres agregar nuevas filas o columnas a un DataFrame existente.

```python
# Crear dos DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9],
                    'B': [10, 11, 12]})
```

```python
# Concatenar DataFrames verticalmente
resultado = pd.concat([df1, df2],axis=0).reset_index(drop=True)
resultado
```

```python
# Concatenar DataFrames horizontalmente (agregar nuevas columnas)
resultado = pd.concat([df1, df2], axis=1).reset_index(drop=True)
resultado
```

2. **Merge**: Para combinar DataFrames utilizando columnas compartidas.

```python
# Crear dos DataFrames con columnas compartidas
df1 = pd.DataFrame({'clave': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
df2 = pd.DataFrame({'clave': ['K2', 'K0', 'K1'], 'B': ['B0', 'B1', 'B2']})

df1
```

```python
df2
```

```python
# Merge basado en la columna 'clave'
resultado = pd.merge(df1, df2, on='clave')
resultado
```

3. **Join**: El método join se utiliza para combinar DataFrames utilizando sus índices.

```python
# Crear dos DataFrames con índices compartidos
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2']}, index=['K0', 'K1', 'K2'])
df2 = pd.DataFrame({'B': ['B0', 'B1', 'B2']}, index=['K0', 'K1', 'K2'])
df1
```

```python
df2
```

```python
# Unir DataFrames basados en los índices
resultado_join = df1.join(df2)
resultado_join
```

`join()` se basa en los índices de los DataFrames, mientras que `merge()` se basa en columnas.

**Profundizando métodos para acceder y seleccionar datos**

**Método .loc[ ]:**
Este método se utiliza para acceder a un grupo de filas y columnas por etiquetas.

Sintaxis: `dataframe.loc[filas, columnas]`


```python
# Crear un DataFrame de ejemplo
datos = {'A': [1, 2, 3, 4],
         'B': [5, 6, 7, 8]}
df = pd.DataFrame(datos, index=['fila1', 'fila2', 'fila3', 'fila4'])
df
```

```python
# Seleccionar una fila por etiqueta de índice
df.loc['fila2']
```

```python
# Seleccionar una fila por etiqueta de índice con columnas específicas
df.loc['fila3', 'A']
```

```python
# Seleccionar un rango de filas por etiquetas de índice
df.loc['fila2':'fila4']
```

```python
# Seleccionar filas basadas en una condición booleana
df.loc[df['A'] > 2]
```

```python
# Modificar valores de una fila seleccionada
df.loc['fila1', 'A'] = 100
df
```

**Método .iloc[ ]:**
Este método se utiliza para acceder a un grupo de filas y columnas por ubicación basada en índices enteros.

Sintaxis: `dataframe.iloc[filas, columnas]`

```python
# Seleccionar la primera fila por índice
df.iloc[0]
```

```python
# Seleccionar una fila y columna específicas por índice
df.iloc[2, 1]
```

```python
# Seleccionar un rango de filas y columnas por índices
df.iloc[1:3, 1]
```

# Numpy aplicado a operaciones en dataframes

```python
import numpy as np
import pandas as pd
```

1. Aplicar una función de NumPy a una columna o DataFrame completo:

```python
# Crear un DataFrame de ejemplo
datos = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(datos)
df
```

```python
# Aplicar la función de NumPy np.sqrt() a una columna
df['Raíz cuadrada'] = np.sqrt(df['A'])
df
```

```python
# Calcular la media y desviación estándar usando NumPy
media = np.mean(df['A'])
desviacion_estandar = np.std(df['A'])
print("Media:", media)
print("Desviación estándar:", desviacion_estandar)
```

2. Realizar operaciones aritméticas entre columnas usando NumPy:

```python
# Crear un DataFrame de ejemplo
datos = {'A': [1, 2, 3, 4, 5],
         'B': [6, 7, 8, 9, 10]}
df = pd.DataFrame(datos)
df
```

```python
# Sumar dos columnas usando NumPy
df['Suma'] = np.add(df['A'], df['B'])
df
```

3. Aplicar una función personalizada de NumPy a un DataFrame completo:

```python
# Crear una función personalizada de NumPy que devuelve un diccionario con nuevas columnas
def funcion_personalizada(x):
    return {'Raíz cuadrada': np.sqrt(x), 'Suma 10': x + 10}

# crear dataframe con resultados
df_resultado = df['A'].apply(funcion_personalizada).apply(pd.Series)
df_resultado
```

```python
#otra manera de crear los resultados anteriores
df['raiz_cuadrada']=df['A'].apply(lambda x: np.sqrt(x))
df['suma_de_10']=df['A'].apply(lambda x: x+10)
df
```

**Evaluando presencia de Valores atipicos**

```python
# Seleccionar solo las variables numéricas
variables_numericas = datos.select_dtypes(include=['int64', 'float64'])

# Crear un diagrama de caja y bigotes múltiples usando Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=variables_numericas, orient='h')
plt.title('Diagrama de Caja y Bigotes para Variables Numéricas del Titanic')
plt.xlabel('Valor')
plt.show()
```

1. Imputacion con diagrama de caja y bigotes

```python
# Crear un diagrama de caja y bigotes múltiples usando Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=datos['Age'], orient='h')
plt.show()
```

```python
# Calcular los cuantiles para 'Age'
q1 = datos['Age'].quantile(0.25)
q3 = datos['Age'].quantile(0.75)
iqr = q3 - q1

# Definir límites para los valores atípicos en la columna 'Age'
lower_limit_age = q1 - 1.5 * iqr
upper_limit_age = q3 + 1.5 * iqr

# Corregir valores de 'Age' que están por encima del límite superior o por debajo del límite inferior
datos['Age'] = np.where(datos['Age'] > upper_limit_age, upper_limit_age,
                        np.where(datos['Age'] < lower_limit_age, lower_limit_age, datos['Age']))

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=datos['Age'], orient='h')
plt.show()
```

2. Filtrar con percentiles

```python
datos.groupby('Pclass')['Fare'].quantile(0.90)
```

```python
percentile_90_fare_by_pclass = datos.groupby('Pclass')['Fare'].quantile(0.90) #definir umbral de percentil 90

above_90th_percentile = datos[datos['Fare'] > datos['Pclass'].map(percentile_90_fare_by_pclass)] #me selecciona outliers

# Plot all data points
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Pclass', y='Fare', data=datos, color='blue', label='All Data')

# Plot fares above the 90th percentile for each passenger class
sns.scatterplot(x='Pclass', y='Fare', data=above_90th_percentile, color='red', label='Above 90th Percentile')

plt.title('Fares Above the 90th Percentile for Each Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks([1, 2, 3])
plt.legend()
plt.show()
```

```python
# Calcular el percentil 90 de la tarifa ('Fare') para cada clase de pasajero ('pclass')
percentil_90_fare_by_pclass = datos.groupby('Pclass')['Fare'].quantile(0.90)

# Filtrar los datos teniendo en cuenta el percentil 90 de la tarifa ('Fare') para cada clase de pasajero ('pclass')
datos_filtrados = datos[datos.apply(lambda x: x['Fare'] <= percentil_90_fare_by_pclass[x['Pclass']], axis=1)]

# Restablecer el índice después de la agrupación
datos_filtrados.reset_index(drop=True, inplace=True)
```

```python
# Plot all data points
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Pclass', y='Fare', data=datos_filtrados, color='blue', label='All Data')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks([1, 2, 3])
plt.legend()
plt.show()
```

```python
max_fare_by_pclass = datos_filtrados.groupby('Pclass')['Fare'].max()

# Imprimir el valor máximo de 'fare' para cada grupo de 'pclass'
print(max_fare_by_pclass)
```

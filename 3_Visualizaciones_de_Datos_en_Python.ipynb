# Matplotlib

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
````

**Matplotlib** es una libreria de Python para visualización de datos que ofrece herramientas poderosas y flexibles para crear una variedad de gráficos.Permite representar datos de manera efectiva en diferentes formas, desde gráficos simples hasta visualizaciones complejas. Su facilidad de uso y calidad en la generación de gráficos la hacen esencial para explorar, analizar y comunicar datos en diversos campos.

**`matplotlib`** es la biblioteca completa que proporciona todas las funcionalidades para crear visualizaciones complejas, mientras que **`matplotlib.pyplot`** es un módulo de nivel superior que facilita la creación rápida de gráficos utilizando una sintaxis simplificada. En muchos casos, matplotlib.pyplot se importa como plt por conveniencia, lo que hace que su uso sea aún más conciso.

````python
plt.subplots() # se utiliza para crear una figura y un conjunto de ejes (subplots) en una sola llamada
````

````python
fig, ax = plt.subplots() #crea una figura (fig) y un conjunto de ejes (ax)
ax.plot([10, 35], [26, 42]) #traza una línea en el gráfico utilizando los ejes ax
````

````python
fig, ax = plt.subplots() #crea una figura (fig) y un conjunto de ejes (ax)
ax.plot([10, 35], [26, 42]) #traza una línea en el gráfico utilizando los ejes ax

ax.set_xlim([0,40]) # desde - hasta donde va el eje X
ax.set_ylim([15,45]) # desde - hasta donde va el eje Y

ax.set_xlabel('eje X') # etiqueta eje x
ax.set_ylabel('eje Y') # etiqueta eje y

ax.set_title('Grafico de línea') # título

ax.axhline(y=26,color='y') #Traza una línea horizontal en el gráfico en el nivel y=26 (eje y), con el color amarillo ('y').
ax.axvline(x=10, color='m') #Traza una línea vertical en el gráfico en el punto x=10 (eje x), con el color magenta ('m').
````

## Grafico de lineas

````python
# Meses del año
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Temperatura promedio mensual en grados Celsius
temperaturas = [18, 19, 21, 24, 27, 29, 31, 31, 29, 26, 22, 19]

# Crear una figura y ejes usando plt.subplots()
fig, ax = plt.subplots()

ax.plot(meses, #eje x
        temperaturas, #eje y
        marker='o' #marcador para cada punto en la grafica
        )

# Etiquetas de los ejes y título
ax.set_xlabel('Mes')
ax.set_ylabel('Temperatura (°C)')
ax.set_title('Temperatura Promedio Mensual')


plt.xticks(rotation=45) # Rotar las etiquetas del eje x para mayor legibilidad

plt.show() # Mostrar el gráfico
````

## Grafico de puntos

````python
# Datos para el eje x (por ejemplo, horas de estudio)
horas_estudio = [2, 3, 4, 5, 6, 7, 8]

# Datos para el eje y (por ejemplo, calificaciones obtenidas)
calificaciones = [65, 70, 75, 80, 85, 90, 95]

# Crear el gráfico de puntos (scatter plot)
plt.scatter(horas_estudio, calificaciones)

# Etiquetas de los ejes y título
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre Horas de Estudio y Calificaciones')

# Mostrar el gráfico
plt.show()
````

## Grafico de barras

````python
# Datos para las barras (por ejemplo, ventas por mes)
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [15000, 18000, 22000, 20000, 23000]

# Crear el gráfico de barras
plt.bar(meses, ventas, color='skyblue')

# Etiquetas de los ejes y título
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.title('Ventas Mensuales')

# Mostrar el gráfico
plt.show()
````

## Grafico circular

````python
categorias = ['Manzanas', 'Plátanos', 'Naranjas', 'Uvas']

# Datos para cada categoría (por ejemplo, cantidad de frutas)
cantidades = [30, 20, 25, 15]

# Colores para cada porción del gráfico
colores = ['lightcoral', 'gold', 'lightskyblue', 'lightgreen']

# Crear el gráfico circular
plt.pie(cantidades, labels=categorias,
        colors=colores,
        autopct='%1.1f%%'
        )

# Título del gráfico
plt.title('Distribución de las Frutas')

# Mostrar el gráfico
plt.show()
````

## Modificando aspectos de graficos usando parametros de matplotlib

````python
x = [0.12, 0.31, 0.45, 0.67, 0.89, 1.23, 1.45, 1.68, 1.78, 2.01]
y = [0.55, 0.67, 0.75, 0.82, 0.93, 1.01, 1.25, 1.37, 1.45, 1.67]

plt.scatter(x, y)

# Etiquetas de los ejes y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
````

````python
# Definir los nuevos parámetros
new_params = {
    'figure.figsize': (16, 4),
    'axes.facecolor': 'lightgrey',  # Color de fondo del área del gráfico
    'figure.facecolor': 'lightblue', # Color de fondo de la figura completa
    'axes.labelcolor' : 'magenta', # color de  las etiquetas de los ejes
    'axes.labelsize' : '15' #tamaño de las etiquetas de los ejes
}

# Modificar los parámetros
mpl.rcParams.update(new_params)
````

## Ejemplo mas avanzado

````python
import yfinance as yf

# Obtener datos de las acciones de Apple y Microsoft
apple_data = yf.download('AAPL', start='2021-01-01', end='2022-01-01')
microsoft_data = yf.download('MSFT', start='2021-01-01', end='2022-01-01')

# Extraer solo los precios de cierre de las acciones para cada empresa
apple_close = apple_data['Close']
microsoft_close = microsoft_data['Close']
````

````python
# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(16, 6))

# Graficar la serie de tiempo de cierre de Apple con estilo minimalista
ax.plot(apple_close, label='Apple', color='navy')

# Graficar la serie de tiempo de cierre de Microsoft con estilo minimalista
ax.plot(microsoft_close, label='Microsoft', color='darkgreen')

# Añadir título y etiquetas con un estilo elegante
ax.set_title('Comparación de Precios de Cierre de Acciones de Apple y Microsoft (2021)', fontsize=12, fontweight='bold')
ax.set_xlabel('Fecha', fontsize=10)
ax.set_ylabel('Precio de Cierre', fontsize=10)
ax.legend(fontsize=12)

# Rotar las etiquetas del eje x para mejor legibilidad
plt.xticks(rotation=45)

# Agregar una rejilla de fondo con líneas más sutiles
plt.grid(color='lightgrey', linestyle='--')

# Cambiar el color de fondo de la figura para un aspecto más elegante
fig.set_facecolor('whitesmoke')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
````

# Seaborn

Seaborn es una especie de "actualización" de Matplotlib para hacer gráficos en Python. Actúa como una capa adicional que embellece tus gráficos de manera más fácil y utiliza Matplotlib en segundo plano para generarlos, lo que significa que son complementarios.

https://seaborn.pydata.org/examples/index.html

````python
import seaborn as sns

import matplotlib.pyplot as plt ## sublibreria de matplotlib
````

````python
penguins = sns.load_dataset("penguins") #esta funcion nos permite tener accesos a ciertos datasets que vienen asociados a la libreria
penguins
````

````python
#HISTOGRAMA

sns.histplot(data=penguins,
             x="flipper_length_mm",
             hue="species" #categorias que quiero tenga en cuenta para diferenciar
             )
plt.show()
````

````python
#HISTOGRAMA (con otro metodo mas flexible)

#mismos parametros que celda anterior con diferencia de que ahora con 'col' se crea una gráfica para cada categoría
sns.displot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            col="species",
            kind="hist" ####probar con kind="kde" para ver diagramas de densidad
            )
plt.show()
````

````python
# GRAFICO DE PUNTOS

sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.show()
````

````python
# GRAFICO DE PUNTOS (con  otro metodo mas flexible)

sns.relplot(data=penguins,
            x="flipper_length_mm",
            y="body_mass_g",
            hue="species",
            col="species",
            kind="scatter"
            )
plt.show()
````

````python
#GRAFICO DE LINEAS

sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species",
             errorbar=None
             )
plt.show()
````

````python
sns.relplot(data=penguins,
            x="flipper_length_mm",
            y="body_mass_g",
            hue="species",
            col="species",
            kind="line",errorbar=None
            )
plt.show()
````

````python
# GRAFICO DE BARRAS

sns.barplot(penguins,
            x="island",
            y="body_mass_g",
            hue="species",
            errorbar=None
            )
plt.show()
````

**Ejemplos mas avanzados: Subplots o sub gráficos**

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Scatterplot 1: Flipper Length vs. Body Mass
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", ax=axes[0])
axes[0].set_title('Flipper Length vs. Body Mass')

# Scatterplot 2: Bill Length vs. Bill Depth
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", ax=axes[1])
axes[1].set_title('Bill Length vs. Bill Depth')

plt.tight_layout()
plt.show()
````

# Adicional: Plotly

Plotly es una biblioteca de visualización interactiva en Python que ofrece gráficos interactivos y dashboards, facilitando la exploración de datos. Sus ventajas incluyen su interactividad, facilidad de uso y variedad de gráficos, aunque puede tener una curva de aprendizaje más pronunciada y depende de la conexión a Internet para algunas funciones avanzadas, en comparación con Matplotlib y Seaborn.

````python
import plotly.express as px
````

**Ejemplo 1**

El conjunto de datos Iris contiene información sobre tres especies de iris (Setosa, Versicolor y Virginica) y medidas de longitud y ancho de sépalos y pétalos.

````python
df1 = px.data.iris()
df1.head()
````

````python
fig = px.scatter_matrix(df1, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")
fig.update_layout(
    width=1000,
    height=600
)
fig.show()
````

**Ejemplo 2**

````python
df2 = px.data.gapminder()
df2.head()
````

````python
fig = px.scatter(df2,
                 x="gdpPercap",  # Producto Interno Bruto per cápita
                 y="lifeExp",  # Expectativa de vida
                 animation_frame="year",
                 animation_group="country",  # para agrupar puntos durante la animación (País)
                 size="pop",  # para determinar el tamaño de los puntos (Población)
                 color="continent",  # para asignar colores a los puntos (Continente)
                 hover_name="country",  # se mostrará al pasar el cursor sobre los puntos (País)
                 facet_col="continent",  # Cada continente en una columna
                 range_x=[100, 40000],
                 range_y=[25, 90],
                 size_max=100)

# Mostrar el gráfico interactivo
fig.show()
````
